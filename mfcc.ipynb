```python
import soundfile as sf
import numpy as np
import os
import glob
import librosa

path = "spoken-language-identification/test/" #path = "spoken-language-identification/train/"
resultsDir= "MFCCtest" #resultsDir = "MFCCtrain"

for filename in glob.glob(os.path.join(path, '*.flac')):
    #loading data
    data, samplerate = sf.read(filename)
    #extracting features
    S = librosa.feature.melspectrogram(data, sr=samplerate, n_mels=128)
    log_S = librosa.power_to_db(S, ref=np.max)
    mfcc = librosa.feature.mfcc(S=log_S, n_mfcc=13)
    mfcc_delta2 = librosa.feature.delta(mfcc, order=2)
    #saving data
    outputFile = os.path.split(filename)[1]
    outputFile = resultsDir + "/" + os.path.splitext(outputFile)[0] +".csv"
    file = open(outputFile, 'w+')
    numpy.savetxt(file, mfcc_delta2, delimiter=",")
    file.close()
    print(outputFile)
```
